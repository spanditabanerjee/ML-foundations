MATRIX INVERSION


```python
import numpy as np
S = np.array([[4,2],[-5,-3]])
S
```

```python
t = np.array([4,-7])
t
```

```python
sinv = np.linalg.inv(S)
sinv
```

```python
w = np.dot(sinv,t)
w
```

```python
#crosschecking formula y = Wx
np.dot(S,w)
```
